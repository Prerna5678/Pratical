```python
import math
import time

class TicTacToe:
    def __init__(self):
        self.board = [[' ' for _ in range(3)] for _ in range(3)]
        # print("TicTacToe board initialized.") # Removed to avoid redundant output during execution

    def display_board(self):
        print("\nCURRENT BOARD:")
        for row in self.board:
            print("|" + "|".join(row) + "|")
        print("\n")

    def make_move(self, row, col, player):
        if 0 <= row < 3 and 0 <= col < 3 and self.board[row][col] == ' ':
            self.board[row][col] = player
            return True
        else:
            return False

    def check_win(self, player):
        # Check rows
        for row in range(3):
            if all(self.board[row][col] == player for col in range(3)):
                return True
        # Check columns
        for col in range(3):
            if all(self.board[row][col] == player for row in range(3)):
                return True
        # Check diagonals
        if all(self.board[i][i] == player for i in range(3)) or \
           all(self.board[i][2 - i] == player for i in range(3)):
            return True
        return False

    def check_draw(self):
        for row in range(3):
            for col in range(3):
                if self.board[row][col] == ' ':
                    return False  # There's an empty spot, not a draw yet
        return True  # Board is full, if no one won, it's a draw

    def evaluate_board(self, ai_player, human_player):
        if self.check_win(ai_player):
            return 10
        elif self.check_win(human_player):
            return -10
        elif self.check_draw():
            return 0
        return 0

    def minimax(self, depth, is_maximizing_player, ai_player, human_player):
        if self.check_win(ai_player):
            return 10 - depth
        if self.check_win(human_player):
            return -10 + depth
        if self.check_draw():
            return 0

        if is_maximizing_player:
            max_eval = -math.inf
            for r in range(3):
                for c in range(3):
                    if self.board[r][c] == ' ':
                        self.board[r][c] = ai_player
                        eval_val = self.minimax(depth + 1, False, ai_player, human_player)
                        self.board[r][c] = ' '
                        max_eval = max(max_eval, eval_val)
            return max_eval
        else:
            min_eval = math.inf
            for r in range(3):
                for c in range(3):
                    if self.board[r][c] == ' ':
                        self.board[r][c] = human_player
                        eval_val = self.minimax(depth + 1, True, ai_player, human_player)
                        self.board[r][c] = ' '
                        min_eval = min(min_eval, eval_val)
            return min_eval

    def find_best_move(self, ai_player, human_player):
        best_eval = -math.inf
        best_move = (-1, -1)

        for r in range(3):
            for c in range(3):
                if self.board[r][c] == ' ':
                    self.board[r][c] = ai_player
                    eval = self.minimax(0, False, ai_player, human_player)
                    self.board[r][c] = ' '

                    if eval > best_eval:
                        best_eval = eval
                        best_move = (r, c)
        return best_move

    def play_game(self, human_player, ai_player):
        current_player = human_player
        print(f"\nGame Start! Human ({human_player}) vs AI ({ai_player})")

        while True:
            self.display_board()

            if self.check_win(ai_player):
                print(f"AI ({ai_player}) wins!")
                break
            elif self.check_win(human_player):
                print(f"Human ({human_player}) wins!")
                break
            elif self.check_draw():
                print("It's a draw!")
                break

            if current_player == human_player:
                print(f"Human's ({human_player}) turn.")
                valid_move = False
                while not valid_move:
                    try:
                        row = int(input("Enter row (0-2): "))
                        col = int(input("Enter column (0-2): "))
                        if self.make_move(row, col, human_player):
                            valid_move = True
                        else:
                            print("Invalid move! Please enter a row and column between 0-2 for an empty cell.")
                    except ValueError:
                        print("Invalid input. Please enter numbers for row and column.")
            else: # AI's turn
                print(f"AI's ({ai_player}) turn...")
                time.sleep(1)
                ai_row, ai_col = self.find_best_move(ai_player, human_player)
                self.make_move(ai_row, ai_col, ai_player)
                print(f"AI plays at ({ai_row}, {ai_col}).")

            current_player = ai_player if current_player == human_player else human_player

# Instantiate the game and start playing
game = TicTacToe()
game.play_game('X', 'O')

print("Game started with human player 'X' and AI player 'O'.")
```
